### Checking the simirality (verb, adj, adv):

```python
import nltk 
from nltk.corpus import wordnet 
 

# compare "load" and "overload:"
# v here denotes the tag verb 
w1 = wordnet.synset('load.v.01')  
w2 = wordnet.synset('overload.v.01') 
print(w1.wup_similarity(w2)) 

# compare "ship" and "boat:"
# n denotes noun
w1 = wordnet.synset('ship.n.01') 
w2 = wordnet.synset('boat.n.01')  
print(w1.wup_similarity(w2))

# Example with adjectives
adj1 = wordnet.synsets('big', pos=wordnet.ADJ)[0]   # Get the first synset of 'big' as an adjective
adj2 = wordnet.synsets('large', pos=wordnet.ADJ)[0] # Get the first synset of 'large' as an adjective

# Calculate similarity between adjectives
adj_similarity = adj1.wup_similarity(adj2)
print(f"Similarity between 'big' and 'large' (adjectives): {adj_similarity}")


# Example with adverbs
adv1 = wordnet.synsets('quickly', pos=wordnet.ADV)[0]  # Get the first synset of 'quickly' as an adverb
adv2 = wordnet.synsets('swiftly', pos=wordnet.ADV)[0]  # Get the first synset of 'swiftly' as an adverb

# Calculate similarity between adverbs
adv_similarity = adv1.wup_similarity(adv2)
print(f"Similarity between 'quickly' and 'swiftly' (adverbs): {adv_similarity}")
```

### Synonyms:

```python
import requests

def get_synonyms(word):
    url = f"https://api.datamuse.com/words?rel_syn={word}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        synonyms = [entry['word'] for entry in data]
        return synonyms
    else:
        print("Error:", response.status_code)
        return []

word = "false"
synonyms = get_synonyms(word)
print("Synonyms for", word + ":", synonyms)
len(synonyms)
```

### Related words:

Here we can change the `limit` and `topics` to find more related words:

```python
import requests

def get_related_words(word, limit=60, topics=None):
    url = f"https://api.datamuse.com/words?ml={word}&max={limit}"
    
    if topics:
        url += f"&topics={topics}"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        related_words = [entry['word'] for entry in data]
        return related_words
    else:
        print("Error:", response.status_code)
        return []

def check_word_in_list(word, word_list):
    if word in word_list:
        return f"'{word}' is in the list."
    else:
        return f"'{word}' is not in the list."



word = "false"
related_words = get_related_words(word, topics="fish, vessel")
word_to_check1 = "inaccurate"
print(related_words, "\n")
print(check_word_in_list(word_to_check1, related_words))

word_to_check2 = "fake"
print(check_word_in_list(word_to_check2, related_words))

word_to_check3 = "incorrect"
print(check_word_in_list(word_to_check3, related_words))
```
