```python
import pandas as pd
# read the excel file
excel_data = pd.read_excel('PIMS Sample Prompts.xlsx')

queries = []
for index, row in excel_data.iterrows():
    # Process each row
    queries.append(row['Prompt'])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Function to scrape search results from Google
def scrape_google_search(query):
    url = f"https://www.google.com/search?q={query}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        search_results = []
        for result in soup.find_all('div', class_='tF2Cxc'):
            title = result.find('h3').text
            link = result.find('a')['href']
            search_results.append({'engine': 'Google', 'title': title, 'link': link})
        return search_results
    else:
       # print(f"Failed to fetch Google search results for query: {query}")
        return []

# Function to scrape search results from Bing
def scrape_bing_search(query):
    url = f"https://www.bing.com/news/search?q={query}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        search_results = []
        for result in soup.find_all('div', class_='t_s'):
            title = result.find('a', class_='title').text
            link = result.find('a', class_='title')['href']
            search_results.append({'engine': 'Bing', 'title': title, 'link': link})
        return search_results
    else:
       # print(f"Failed to fetch Bing search results for query: {query}")
        return []

# Queries for test
queries = [
    "Vessel caught misreporting catch amount",
    "Vessel caught falsifying fishing logs",
    "Vessel caught with incorrect catch reports",
    "Vessel caught falsifying catch during transport",
    "Vessel caught misreporting catch for fishing competition",
    "Vessel caught misreporting catch for quota manipulation"
]

# List of search engines to scrape
search_engines = {
    'Google': scrape_google_search,
    'Bing': scrape_bing_search,
}

# Data storage
data = []

# Loop through each query and each search engine
for query in queries:
    for engine_name, scrape_function in search_engines.items():
        print(f"Scraping {engine_name} for query: {query}")
        results = scrape_function(query)
        
        if results:
            for result in results[:20]:  # Limiting to the top 20 results per engine
                data.append({
                    'Query': query,
                    'Engine': engine_name,
                    'Title': result['title']
                })
        else:
            data.append({
                'Query': query,
                'Engine': engine_name,
                'Title': 'No results'
            })
        
        time.sleep(1)  # Adding a delay to be respectful of website policies

# Create DataFrame
df = pd.DataFrame(data)

# Print the table
print(df.head(10))  # Adjust the number as needed
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# Initialize the vectorizer
vectorizer = TfidfVectorizer()

# Combine all text data (queries and content) for consistent vectorization
all_texts = df['Query'].tolist() + df['Title'].tolist()

# Fit the vectorizer on all texts
vectorizer.fit(all_texts)

# Transform queries and content separately
query_vectors = vectorizer.transform(df['Query'])
content_vectors = vectorizer.transform(df['Title'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity
relevance_scores = []

for i in range(len(df)):
    query_vector = query_vectors[i]
    content_vector = content_vectors[i]
    relevance_score = cosine_similarity(query_vector, content_vector)[0][0]
    relevance_scores.append(relevance_score)

# Add relevance scores to the DataFrame
df['relevance_score'] = relevance_scores

# Calculate precision
# Calculate precision per query and engine
precision_values = []
for (query, engine), group in df.groupby(['Query', 'Engine']):
    total_documents = len(group)
    relevant_documents = len(group[group['relevance_score'] > 0])
    
    if total_documents > 0:
        precision = relevant_documents / total_documents
    else:
        precision = 0.0
    
    precision_values.extend([precision] * total_documents)  # Extend precision to match group size

# Add precision to DataFrame
df['precision'] = precision_values

# Display DataFrame with relevance score and precision
df.head(13)
```

## SOME VISUALIZATIONS

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Plotting using seaborn
plt.figure(figsize=(6, 3))
sns.scatterplot(data=df, x='relevance_score', y='precision', hue='Engine', style='Engine', s=100)
plt.title('Precision vs. Relevance Score by Search Engine')
plt.xlabel('Relevance Score')
plt.ylabel('Precision')
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.legend(title='Engine', loc='best')
plt.grid(True)
plt.show()
```

```python
# Calculate mean precision per engine and query
mean_precision = df.groupby(['Query', 'Engine'])['precision'].mean().unstack()

# Plotting
mean_precision.plot(kind='bar', figsize=(12, 6))
plt.title('Mean Precision by Query and Search Engine')
plt.xlabel('Query')
plt.ylabel('Mean Precision')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Engine', loc='upper right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Create a pairplot using seaborn
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='Engine', vars=['relevance_score', 'precision'], diag_kind='hist', markers=['o', 's'])
plt.suptitle('Pairplot of Relevance Score vs. Precision by Search Engine', y=1.02)
plt.show()
```
