#### Computing the average days for each engine

```python
# Set the date words:
from datetime import datetime

hour = ['hour', 'hours']
day = ['day', 'days']
week = ['week', 'weeks']
month = ['month', 'months']
year = ['year', 'years']
# consider 1 week = 7 days, 1 month = 30 days, 1 year = 365 days for convenience

def convert_day(text):
    # In some cases, dates are given as YYYY-MM-DD
    pattern = r'^\d{4}-\d{2}-\d{2}$'
    if re.match(pattern, text):
        date = datetime.strptime(text, "%Y-%M-%d")
        today = datetime.combine(date.today(), datetime.min.time())
        days = str(today - date).split()
        return int(days[0])
    txtlist = text.split()
    if txtlist[1] in ['hour', 'hours']:
        return 1
    if txtlist[1] in ['day', 'days']:
        return int(txtlist[0])
    if txtlist[1] in ['week', 'weeks']:
        return int(txtlist[0])*7
    if txtlist[1] in ['month', 'months']:
        return int(txtlist[0])*30
    if txtlist[1] in ['year', 'years']:
        return int(txtlist[0])*365

        
```

```python
# Computing Google average
googledays = 0 
google_days_list = []

for search_result in google_search:
    tmp_list = []
    for index, result in enumerate(search_result, start=1):
        tmp = convert_day(result['date'])
        googledays += tmp
        tmp_list.append(tmp)
    google_days_list.append(tmp_list)

google_average_days = googledays/len(google_search)
```

```python
# Computing Yahoo average
yahoodays = 0 
yahoo_days_list = []

for search_result in yahoo_search:
    tmp_list = []
    for index, result in enumerate(search_result, start=1):
        tmp = convert_day(result['date'])
        yahoodays += tmp
        tmp_list.append(tmp)
    yahoo_days_list.append(tmp_list)

yahoo_average_days = yahoodays/len(yahoo_search)
```

```python
# Computing Bing average
bingdays = 0 
bing_days_list = []

for search_result in bing_search:
    tmp_list = []
    for index, result in enumerate(search_result, start=1):
        tmp = convert_day(result['date'])
        bingdays += tmp
        tmp_list.append(tmp)
    bing_days_list.append(tmp_list)

bing_average_days = bingdays/len(bing_search)
```

```python
#print(google_average_days, yahoo_average_days, bing_average_days)

search_stats['Google days'] = google_days_list
search_stats['Yahoo days'] = yahoo_days_list
search_stats['Bing days'] = bing_days_list

search_stats.head(20)
```
